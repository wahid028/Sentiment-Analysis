```python
import pandas as pd
```

```python
df = pd.read_csv('../content/train.csv')
```

```python
df.head()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df['sentiment'].value_counts()
```

```python
import nltk 
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
import string
import re
```

```python
#Remove punctuations
# string.punctuation
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

df['text_punct'] = df['text'].apply(lambda x: remove_punct(x))
df.head()
```

```python
#Tokenization
def tokenization(text):
    text = re.split('\W+', text)
    return text

df['text_tokenized'] = df['text_punct'].apply(lambda x: tokenization(x.lower()))
df.head()
```

```python
#Remove stopwords
stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
df['text_nonstop'] = df['text_tokenized'].apply(lambda x: remove_stopwords(x))
df.head()
```

```python
#Stemming
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

df['text_stemmed'] = df['text_nonstop'].apply(lambda x: stemming(x))

#Lammitization
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return " ".join(text)

df['text_lemmatized'] = df['text_nonstop'].apply(lambda x: lemmatizer(x))
df.head()
```

```python
df_new = df[['text_lemmatized', 'sentiment']]
df_new.head()
```

```python
df_new.rename(columns = {'text_lemmatized':'clean_text'}, inplace = True)
df_new.head()
```

```python
!pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def vader_scores(feedbacktext, category):
    return vader.polarity_scores(feedbacktext).get(category)

vader = SentimentIntensityAnalyzer()

df_new["Negative_Score"] = df_new.apply(lambda row : vader_scores(df_new["clean_text"][row.name], "neg"),axis=1)
df_new["Neutral_Score"] = df_new.apply(lambda row : vader_scores(df_new["clean_text"][row.name], "neu"),axis=1)
df_new["Positive_Score"] = df_new.apply(lambda row : vader_scores(df_new["clean_text"][row.name], "pos"),axis=1)
df_new["Compound_Score"] = df_new.apply(lambda row : vader_scores(df_new["clean_text"][row.name], "compound"),axis=1)

#Function for addressing polarity
def getAnalysis(score):
  if score <= -0.05:
    return 'negative'
  elif -0.05 < score < 0.05:
    return 'neutral'
  else:
    return 'positive'

df_new['VADER_Analysis'] = df_new['Compound_Score'].apply(getAnalysis )
```

```python
df_new.head()
```

```python
# Import accuracy_score to check performance
from sklearn.metrics import accuracy_score
# Compare Actual and Predicted
print("VADER accuracy score: ", accuracy_score(df_new['sentiment'],df_new['VADER_Analysis']))
```

```python
from textblob import TextBlob
```

```python
#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create new columns ‘Subjectivity’
df_new['TextBlob_Subjectivity'] = df_new['clean_text'].apply(getSubjectivity)

#Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

#Create new columns ‘Polarity’
df_new['TextBlob_Polarity'] = df_new['clean_text'].apply(getPolarity)

#Function for addressing polarity
def getAnalysis(score):
  if score < 0:
    return 'negative'
  elif score == 0:
    return 'neutral'
  else:
    return 'positive'

df_new['TextBlob_Analysis'] = df_new['TextBlob_Polarity'].apply(getAnalysis)
```

```python
# Compare Actual and Predicted
print("TextBlob accuracy score: ", accuracy_score(df_new['sentiment'],df_new['TextBlob_Analysis']))
```
